```python
import pandas as pd
import googlemaps

# Load your data into a Pandas DataFrame
df = pd.read_csv('/content/Simple.csv')

# Initialize Google Maps client with your API key
api_key = 'AIzaSyB6h-Q6ebaWf_xV2JyMdp-nEmCUK6lI6iI'
gmaps = googlemaps.Client(api_key)

# Create new columns for rooftop northeast and southwest coordinates
df['rooftop_northeast_latitude'] = None
df['rooftop_northeast_longitude'] = None
df['rooftop_southwest_latitude'] = None
df['rooftop_southwest_longitude'] = None

# Iterate over each row
for index, row in df.iterrows():
    locality = row['Locality']

    # Skip iteration if locality is empty
    if pd.isna(locality):
        print(f'Skipping empty row at index {index}')
        continue

    # Geocode to retrieve place_id
    geocode_result = gmaps.geocode(locality)

    if geocode_result:
        place_id = geocode_result[0]['place_id']

        # Use place_id to get place details
        place_details = gmaps.place(place_id, fields=['geometry'])

        if 'geometry' in place_details['result']:
            geometry = place_details['result']['geometry']

            if 'viewport' in geometry:
                viewport = geometry['viewport']
                northeast = viewport['northeast']
                southwest = viewport['southwest']

                # Extract rooftop northeast and southwest coordinates
                rooftop_northeast_latitude = northeast['lat']
                rooftop_northeast_longitude = northeast['lng']
                rooftop_southwest_latitude = southwest['lat']
                rooftop_southwest_longitude = southwest['lng']

                # Save the rooftop coordinates in the DataFrame
                df.at[index, 'rooftop_northeast_latitude'] = rooftop_northeast_latitude
                df.at[index, 'rooftop_northeast_longitude'] = rooftop_northeast_longitude
                df.at[index, 'rooftop_southwest_latitude'] = rooftop_southwest_latitude
                df.at[index, 'rooftop_southwest_longitude'] = rooftop_southwest_longitude

                print(f'Rooftop coordinates found for "{locality}"')
            else:
                print(f'Viewport information not available for "{locality}"')
        else:
            print(f'Geometry information not available for "{locality}"')
    else:
        print(f'Location not found for "{locality}"')

# Save the DataFrame with rooftop coordinates to a new CSV file
output_file = 'rooftop_coordinates.csv'
df.to_csv(output_file, index=False)
print(f'Task completed. Rooftop coordinates saved to "{output_file}".')
```
